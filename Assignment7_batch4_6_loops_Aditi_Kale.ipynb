## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
# Write your answer here
Example=[[10,20], [30], [5, 7, 8]]
count = 0
for item in Example:
    for num in item:
        count+=num
print("Sum of All Numbers=",count)
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
# Write your answer here
numbers = [[10, -20, 30], [5, 15, -25], [0, 40, -10]]
for i in numbers:
    for num in range(len(i)):
        if i[num]<0:
            i[num]=0
numbers
````

## **19. Find total stock by summing all 'quantity' values in a nested dictionary.**

````python
# Write your answer here
total_stock=0
stock = {'P1': {'name': 'Pen', 'quantity': 50},'P2': {'name': 'Notebook', 'quantity': 75},'P3': {'name': 'Eraser', 'quantity': 30}}
for num, detail in stock.items():
    for name, quan in detail.items():
        if name=="quantity":
            total_stock += quan
print(total_stock)
````
