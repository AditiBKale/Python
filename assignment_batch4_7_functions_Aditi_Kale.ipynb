### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
def adapter(operations):
    for i in operations:
        op = i["op"]
        a = i["a"]
        b = i["b"]
        if op=="add":
            print(a+b)
        elif op=="mul":
            print(a*b)
        elif op=="sub":
            print(a-b)
        else:
            print(a/b)

operations=[{"op":"add","a":10,"b":2},{"op":"mul","a":10,"b":2}]

adapter(operations)

# def adapter(operations):
#     for i in operations:
#         if i["op"]=="add":
#             print(i["a"]+i["b"])
#         elif i["op"]=="mul":
#             print(i["a"]*i["b"])
#         elif i["op"]=="sub":
#             print(i["a"]-i["b"])
#         else:
#             print(i["a"]/i["b"])

# operations=[{"op":"add","a":10,"b":2},{"op":"mul","a":10,"b":2}]
# adapter(operations)
```

### Write a function process_students(data) that prints total marks and grade for each student.
- Example: [ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
- Expected Output: Name: A | Total: 240 | Grade: A Name: B | Total: 200 | Grade: B

```python
def process_students(data):
    for i in data:
        name = i["name"]
        marks = i["marks"]
        total = sum(marks)

        if total>=240:
            grade = "A"
        elif total>=200:
            grade = "B"
        else:
            grade = "None"

        print(f"Name:{name}|Total:{total}|Grade:{grade}")

data =[{"name":"A", "marks":[80,90,70]}, 
       {"name":"B", "marks":[60,75,65]}] 

process_students(data)
```

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_discounts(products):
    for i in products:
        name = i["name"]
        price = i["price"]
        discount = i["discount"]
        discount_price = price - (price*discount/100)
        price = int(discount_price)
        print(f"name:{name},price:{price},discount:{discount}")

products=[{"name":"Laptop","price":50000,"discount":10},{"name":"Phone","price":20000,"discount":5}]

apply_discounts(products)
```

### Write a function generate_report(employees) that calculates bonus based on salary slabs. 
- Slabs: 
- salary > 50,000 → bonus = 15%, 
- salary 30,000–50,000 → bonus = 10%, 
- salary < 30,000 → bonus = 5%. 

#### Example employees = 
- [ {"id": 101, "name": "Amit", "age": 28, "department": "IT", "salary": 55000}, {"id": 102, "name": "Neha", "age": 24, "department": "HR", "salary": 40000}, {"id": 103, "name": "Raj", "age": 31, "department": "Finance", "salary": 65000}, {"id": 104, "name": "Simran", "age": 29, "department": "IT", "salary": 60000}, {"id": 105, "name": "Vijay", "age": 35, "department": "Admin", "salary": 35000} ]

```python
def generate_report(employees):
    for i in employees:
        id = i["id"]
        name = i["name"]
        age = i["age"]
        department = i["department"]
        salary = i["salary"]

        if salary>50000:
            bonus = 15/100
        elif salary>=30000 and salary<=50000:
            bonus = 10/100
        elif salary>30000:
            bonus = 5/100
        else:
            print("none")
        total_salary = int(salary+(salary*bonus))

        print(f"id:{id},name:{name},age:{age},department:{department},Salary:{total_salary}")

employees =[{"id": 101, "name": "Amit", "age": 28, "department": "IT", "salary": 55000},
            {"id": 102, "name": "Neha", "age": 24, "department": "HR", "salary": 40000},
            {"id": 103, "name": "Raj", "age": 31, "department": "Finance", "salary": 65000},
            {"id": 104, "name": "Simran", "age": 29, "department": "IT", "salary": 60000},
            {"id": 105, "name": "Vijay", "age": 35, "department": "Admin", "salary": 35000} ]

generate_report(employees)
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
def out_of_stock(items):
    for i in items:
        name = i["name"]
        stock = i["stock"]
        if stock<20:
            print(f"name:{name} stock:{stock}")

items = [{"name": "Milk", "stock": 10},
    {"name": "Bread", "stock": 25},
    {"name": "Eggs", "stock": 18},
    {"name": "Butter", "stock": 5},
    {"name": "Cheese", "stock": 40},
    {"name": "Soap", "stock": 12},
    {"name": "Shampoo", "stock": 30}]

out_of_stock(items)
```

### Write total_inventory_value(items) → sum(price * stock).

### Write search_product(items, name) → return matching product(s).

```python
def search_product(items,name):
    result=[]
    for i in items:
        if name.lower() in i['name'].lower():
            result.append(i)
    print(result)

items = [
    {"id": 1, "name": "Notebook", "price": 50, "stock": 15},
    {"id": 2, "name": "Pen", "price": 10, "stock": 120},
    {"id": 3, "name": "Marker", "price": 30, "stock": 8},
    {"id": 4, "name": "Eraser", "price": 5, "stock": 25},
    {"id": 5, "name": "Stapler", "price": 80, "stock": 12}
]

search_product(items,'pen')
```
