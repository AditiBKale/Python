## Q1: Create a class called Student

Requirements:
- Class variable: `school_name = "ABC International School"`
- Constructor: takes `name` and `marks`
- Instance method: `display_details()`
- Class method: `change_school()`
- Static method: `is_pass(marks)` returns "Pass" if marks â‰¥ 35 else "Fail"


```python
class student:
    school_name = "ABC International School"
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    def display_details(self):
        print("Name:",self.name)
        print("Marks:",self.marks)
        print("School Name:",student.school_name)
    @classmethod
    def change_school(cls,new_school):
        cls.school_name = new_school
        
    def is_pass(marks):
        if marks>=35:
            print("Pass")
        else:
            print("Fail")

obj = student("Aditi", 90)
obj.display_details()

student.is_pass(90)

student.change_school("PQR International School")
obj.display_details()
```

## Q2: Create a class Calculator

Requirements:
- Class variable: `count_operations = 0`
- Instance method: `add(a, b)`
- Class method: `increment_count()`
- Static method: `info()`


```python
class Calculator:
    count_operations = 0
    def add(self,a,b):
        result = a+b
        Calculator.increment_count()
        return result
    @classmethod
    def increment_count(cls):
        cls.count_operations+=1
    @staticmethod
    def info():
        print("Basic Operations")

obj = Calculator()
print(obj.add(10,20))
print(obj.add(40,50))

print(Calculator.count_operations)
Calculator.info()



        
```

- Class variable: `bank_name = "SBI"`
- Constructor: `name`, `balance`
- Instance method: `deposit(amount)`
- Class method: `change_bank(new_name)`
- Static method: `validate_amount(amount)`


```python
class BankAccount:
    bank_name = "SBI"
    def __init__(self,name, balance):
        self.name = name
        self.balance = balance
    def deposite(self,amount):
        if BankAccount.validate_amount(amount):
            self.balance+=amount
        else:
            print("Invalid Amount")
    @classmethod
    def change_bank(cls,new_name):
        cls.bank_name = new_name
    @staticmethod
    def validate_amount(amount):
        return amount>0

obj = BankAccount("Aditi", 2000)
obj.deposite(3000)
print(obj.balance)
BankAccount.change_bank("BOI")
print(BankAccount.bank_name)

        
```
